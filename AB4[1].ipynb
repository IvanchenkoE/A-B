```python
df.groupby('con_treat').agg({'converted':'mean'})
```

```python
# конверсия в тестовой группе уменьшилась
```

```python
tst=pd.crosstab(df.con_treat, df.converted)
tst
```

```python
stats.chi2_contingency(tst) # уровень значимости 0.05
```

```python
# не можем отклонить Н0, что переменные con_treat и converted не связаны между
# собой
```

```python
df_m=pd.merge(df, df_, left_on='id', right_on='id')
df_m
```

```python
df_m.groupby(['country', 'con_treat']).agg({'converted':'mean'})
```

```python
# конверсия увеличилась только в UK
```

```python
df_uk=df_m[df_m.country=='UK']
df_uk
```

```python
tst_uk=pd.crosstab(df_uk.con_treat, df_uk.converted)
tst_uk
```

```python
stats.chi2_contingency(tst_uk) # уровень значимости 0.05
```

```python
# не можем отклонить Н0, что переменные con_treat и converted не связаны между
# собой
```

```python
cnt=[tst_uk.iloc[0,1], tst_uk.iloc[1,1]]
```

```python
nds=[tst_uk.iloc[0,0]+tst_uk.iloc[0,1], tst_uk.iloc[1,0]+tst_uk.iloc[1,1]]
```

```python
proportions_ztest(cnt, nds)
```

```python
# по z тесту для пропорций, также не можем отклонить Н0, нет различий между
# пропорциями в двух группах в UK
```
